```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
labels = data['price']
conv_dates = [1 if values == 2014 else 0 for values in data.date ]
data['date'] = pd.to_datetime(data['date'])
data['date'] = data['date'].dt.year.apply(lambda x: 1 if x == 2014 else 0)
train1 = data.drop(['id', 'price'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.20,random_state =2)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

y_pred_lr = reg.predict(x_test)

print("R² Score:", reg.score(x_test, y_test))
print("MAE: $", round(mean_absolute_error(y_test, y_pred_lr), 2))
print("RMSE: $", round(np.sqrt(mean_squared_error(y_test, y_pred_lr)), 2))
```
